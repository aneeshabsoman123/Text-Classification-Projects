# corpora and corpus samples distributed with NLTK

**webtext and nps_chat** are files inside nltk.corpus which containes sentences

**Brown Corpus** corpus contains text from 500 sources, and the sources have been categorized by genre, such as news, editorial
It is a convenient resource for studying systematic differences between genres, a kind of linguistic inquiry known as stylistics





```python
from nltk.corpus import brown
nltk.download('brown')

brown.categories()
```

```python
news_text = brown.words(categories='news') 
fdist = nltk.FreqDist([w.lower() for w in news_text])
fdist
```

```python
modals = ['can', 'could', 'may', 'might', 'must', 'will']

for index,m in enumerate(modals):
    print(index,m)
    
for m in modals:
    print (m + ':', fdist[m])
```
