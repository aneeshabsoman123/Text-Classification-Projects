# 2.BIGRAMS

```python
nltk.bigrams(['more', 'is', 'said', 'than', 'done'])
```

nltk.bigrams() returns an iterator (a generator specifically) of bigrams. If you want a list, pass the iterator to list(). It also expects a sequence of items to generate bigrams from, so you have to split the text before passing it 

```python
text = "When I first met her she was very quiet. She remained quiet during the entire two hour long journey from Stony Brook to New York."

#TEXT is split and passed into bigram function
#the output is made to form a list
bigrm = list(nltk.bigrams(text.split()))

#calling a map(function,iterable)
#function is joining the words
#iterable is the list of bigrm
a=map(' '.join, bigrm)

#map output is seperated by comman and stored in list
print(list(a), sep=', ')
```

**method 1 to generate n-grams**

```python
#function for generating n-grams
import re

def generate_ngrams(text,n):

    # split sentences into tokens
    tokens=re.split("\\s+",text)
    ngrams=[]

    # collect the n-grams
    for i in range(len(tokens)-n+1):
       temp=[tokens[j] for j in range(i,i+n)]
       ngrams.append(" ".join(temp))

    return ngrams
```

```python
generate_ngrams(text2,10)
```
