# Synonymns through Word net

NLTK includes the English WordNet, with 155,287 words and 117,659 synonym sets.

```python
from nltk.corpus import wordnet as wn
wn.synsets('motorcar')
```

motorcar has just one possible meaning and it is identified as car.n.01, the first noun sense of car. The entity car.n.01 is called a synset, or synonym set, a collection of synonymous words 

```python
#finding synonymn of motor car
wn.synset('car.n.01').lemma_names()  #wordnet object is changed to function hence () needed
```

```python
wn.synset('car.n.01').definition()
```

```python
wn.synset('car.n.01').examples
```

```python
wn.synset('car.n.01').lemmas()
```

```python
wn.lemma('car.n.01.automobile').name()
```

# Hyponymy

**finding the hyponyms of cars**

```python
motorcar = wn.synset('car.n.01')
types_of_motorcar = motorcar.hyponyms()
a=[lemma.name() for synset in types_of_motorcar for lemma in synset.lemmas()]
a
```

**finding the hypernyms of cars**

```python
motorcar.hypernyms()
motorcar = wn.synset('car.n.01')
types_of_motorcar = motorcar.hypernyms()
b=[lemma.name() for synset in types_of_motorcar for lemma in synset.lemmas()]
b
```

# meronyms

```python
tree = wn.synset('tree.n.01')
meronym_of_tree = tree.part_meronyms()
c=[lemma.name() for synset in meronym_of_tree for lemma in synset.lemmas()]
c
```
