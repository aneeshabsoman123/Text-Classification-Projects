```python
def find_pattern(text,patterns):
    if re.search(patterns,text):
       return re.search(patterns,text)
    else:
       return 'Not found'
```

# Quantifiers

```python
print(find_pattern('ac','ab?'))
print(find_pattern('abc','ab?'))
print(find_pattern('abbc','ab?'))
```

```python
print(find_pattern('99956714948129861194','8129861194?'))
```

```python
print(find_pattern('ac','ab+'))
print(find_pattern('abc','ab+'))
print(find_pattern('abbc','ab+'))
```

```python
print(find_pattern('ac','ab*'))
print(find_pattern('abc','ab*'))
print(find_pattern('abbc','ab*'))
```

```python
# matches if a charachter is present exactly n number of times
print(find_pattern('abbc','ab{2}'))
print(find_pattern('abbbbbc','ab{2}'))
```

```python
print(find_pattern('abbbbbbbc','ab{3,5}'))
print(find_pattern('abbbbbbbbbbc','ab{6,7}'))
```

```python
def match_pattern(text,patterns):
    if re.match(patterns,text):
        return re.match(patterns,text)
    else:
        return('Not found')
```

```python
def find_pattern(text,patterns):
    if re.search(patterns,text):
       return re.search(patterns,text)
    else:
       return 'Not found'
```

```python
print(find_pattern('abbc','b+'))
print(match_pattern('abbc','b+'))
```

```python
print(match_pattern('bbc','b+'))
```
